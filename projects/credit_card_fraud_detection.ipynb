```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#https://www.kaggle.com/datasets/mlg-ulb/creditcardfraud?resource=download

data=pd.read_csv('creditcard.csv')
```

```python
data.head()
```

```python
data.tail()
```

```python
data.info()
```

```python
#checking the number of missing values in each column
data.isnull().sum()
```

```python
#distribution of legit transaction fraudlent transation

data['Class'].value_counts()

#highly unbalanced
```

```python
legit = data[data.Class==0]
fraud = data[data.Class==1]

print(fraud.shape)
print(legit.shape)
```

```python
#statistical measures of data

legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transaction

data.groupby('Class').mean()
```

```python
# Undersmapling

#Build a sample dataset containing similiar distribution of normal transaction and Fraudlent transaction

```

```python
legit_sample=legit.sample(n=492)
```

```python
#COncatenating two dataframes
```

```python
new_data=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_data.head()
```

```python
new_data.tail()
```

```python
new_data['Class'].value_counts()
```

```python

new_data.groupby('Class').mean()
```

```python
x=new_data.drop(columns='Class',axis=1)
Y=new_data['Class']
```

```python
print(x)
```

```python
print(Y)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
#Model evaluation
```

```python
x_train_pred=model.predict(X_train)
```

```python
training_acc_score= accuracy_score(x_train_pred,Y_train)
```

```python
print(training_acc_score)
```

```python
x_test_pred=model.predict(X_test)
```

```python
test_acc_score= accuracy_score(x_test_pred,Y_test)
```

```python
print(test_acc_score)
```
